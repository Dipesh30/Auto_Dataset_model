Droppig the carlength and car width column

```python
data.drop(['carlength','carwidth'],axis=1,inplace=True)
```

Label encoding columns

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['make'] = le.fit_transform(data['make'])
data['carbody'] = le.fit_transform(data['carbody'])
data['enginetype'] = le.fit_transform(data['enginetype'])
data['cylindernumber'] = le.fit_transform(data['cylindernumber'])
data['fuelsystem'] = le.fit_transform(data['fuelsystem'])
```

One hot encoding binary columns

```python
cat_col = ['fueltype','aspiration','doornumber','enginelocation', 'drivewheel']
data = pd.get_dummies(data, columns=cat_col, drop_first=True)
```

```python
data.head()
```

Scaling the numerical columns

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
num_col = ['car_area','carheight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']
data[num_col] = sc.fit_transform(data[num_col])
```

```python
data
```

Spiltting the Dataset into feature and target variables

```python
X = data.drop(['price'],axis=1)
y = data[['price']]
```

Changing the shape target of the column

```python
y = y.values.ravel()
```

Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

Finding the Best Model

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 100, max_features = 'auto', max_depth = 6, min_samples_split = 2,min_weight_fraction_leaf = 0.0)
rfr.fit(X_train, y_train)
print(rfr.score(X_train,y_train))
print(rfr.score(X_test,y_test))
```

Grid Search for better accuracy

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [3,4,5,6,7,8],
    'max_features': ['auto'],
    'min_samples_split': [1,2,3,4],
    'n_estimators': [85,90,95,100,105,110,115,120],
    'min_weight_fraction_leaf': [0.0,0.1,0.4]
}
rf = RandomForestRegressor()
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid,cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(X,y)
grid_search.best_params_
```
